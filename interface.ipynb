```python
def predict_all_images(fish_img_path):
    import pickle
    import os
    from PIL import Image
    import numpy as np


    # save the iris classification model as a pickle file
    model_pkl_file = "pretrained_model.pkl"  

    # load model from pickle file
    with open(model_pkl_file, 'rb') as file:  
        fetched_model = pickle.load(file)
    
    # create voting list
    y_predict = []
    counter = 0
    
    # load images
    for images_name in os.listdir(fish_img_path):
        if (counter >= 20):
            break
        if (images_name.endswith(".png")):
            try:
                png_image = Image.open (fish_img_path + "/" + images_name)
            except:
                continue
        image_array = np.array(png_image)
        image_array = np.expand_dims(image_array, axis=0)
        
        # evaluate model 
        single_prediction = fetched_model.predict(image_array, verbose=0)
        predicted_class_index = np.argmax(single_prediction)
        y_predict.append(predicted_class_index)
        class_labels = ["Pomfret", "SilverCarp", "Tincaeus"]  
        predicted_label = class_labels[predicted_class_index]

        # check results
        print(counter, predicted_label)
        counter += 1
    
    prediction_count = np.bincount(y_predict)    
    predicted_class_index = np.argmax(prediction_count)
    class_labels = ["Pomfret", "SilverCarp", "Tincaeus"] 
    final_decision = class_labels[predicted_class_index]
    # print("Final decision: ", final_decision)
    return final_decision
```

```python
def capture_image(write_path):
    import cv2

    cap = cv2.VideoCapture(0) # video capture source camera (Here webcam of laptop) 
    ret,frame = cap.read() # return a single frame in variable `frame`

    cv2.imshow('img1',frame) #display the captured image
    cap.release()

    # Save the captured frame as an image
    cv2.imwrite(write_path, frame)

    # Release the camera
    cap.release()
    
    return ret, frame
```

```python
import tkinter as tk

def clear_directory(directory):
    import os
    # Iterate over all files in the directory
    for filename in os.listdir(directory):
        # Get the full path of the file
        file_path = os.path.join(directory, filename)
        # Check if the path is a file
        if os.path.isfile(file_path):
            # Delete the file
            os.remove(file_path)
        # If the path is a directory, recursively call clear_directory
        elif os.path.isdir(file_path):
            clear_directory(file_path)

def window_creation():
    import os
    global root
    global photographer_counter
    global window_counter
    global directory_path
    
    photographer_counter = 0
    # Create a Tkinter window
    root = tk.Tk()
    root.title("Hello fisher")
    root.geometry("1000x1000+10+10")  # Set the size of the window
  
    # Make one diretory for storing new photos
    directory_path = "Fish" + str(window_counter)
    try:
        os.mkdir(directory_path)
    except:
        clear_directory(directory_path)
        
    # Create a label widget to display the captured image
    text_displayer = tk.Label(root, text = previous_fish_species)
    text_displayer.pack(side="top")
    
    # Create a button widget
    photo_button = tk.Button(root, text="Take Picture", command=photo_button_click)
    predict_button = tk.Button(root, text="Start predicting", command=predict_button_click)

    photo_button.pack(side=tk.LEFT, padx=10, pady=10, anchor="sw")
    predict_button.pack(side=tk.RIGHT, padx=10, pady=10, anchor="se")
    

    # Update
    window_counter += 1

def photo_button_click():
    # TODO: dealing with photographing button
    import tkinter as tk
    from PIL import Image, ImageTk
    import cv2
    
    global photographer_counter
    
    import cv2
    cap = cv2.VideoCapture(0)     
    ret,frame = cap.read() 
    if ret:
        write_path = directory_path + "/fish_imgs" + str(photographer_counter) + ".png"   
        # frame_redo = cv2.convertScaleAbs(frame, alpha=1) 
        frame_redo_store = cv2.resize(frame, (648, 486))
        frame_redo_display = cv2.resize(frame, (200, 150))
        cv2.imwrite(write_path, frame_redo_store)
        # Convert the resized frame to an ImageTk format
        pil_image = Image.fromarray(frame_redo_display)
        tk_image = ImageTk.PhotoImage(image=pil_image)
        
        # Update the image displayed in the label widget
        image_displayer = tk.Label(root, image = tk_image)
        image_displayer.pack(side="top", anchor="center")
    
        # cv2.imwrite(write_path, frame_resized_train_rgb)
        
        photographer_counter += 1  # one more image taken
        cap.release()
        
def predict_button_click():
    # TODO: dealing with prediction
    global previous_fish_species
    previous_fish_species = "Previous fish:" + predict_all_images(directory_path)
    root.destroy()  # Destroy the current window
    window_creation() # Create a window again

window_counter = 0
previous_fish_species = "Nothing Caught Yet. Good Luck!"
window_creation()

# Run the Tkinter event loop
root.mainloop()
```
